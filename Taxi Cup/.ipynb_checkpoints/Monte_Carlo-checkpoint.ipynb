# Wrapper

```python
import gym

class DiagonalTaxiWrapper(gym.Wrapper):
    """
    A wrapper class for the Taxi-v3 environment that allows diagonal movement.
    """
    
    def __init__(self, env):
        super().__init__(env)
         
        # Modify the action space here
        self.action_space = spaces.Discrete(10)  # Restrict to three actions
    
        
    def step(self, action):
        # Convert the action to a tuple of (delta_x, delta_y).
        if action == 0 or action == 1 or action ==2 or action == 3 or action == 4 or action == 5:
            act = self.env.step(action)
            return act
        elif action == 6:
            action0 = 0
            action1 = 2
           
            act = self.env.step(action0)
            act = self.env.step(action1)
            return act 

            
        elif action == 7:
             # Move diagonally top-right.
            action0 = 1
            action1 = 2
           
            act = self.env.step(action0)
            act = self.env.step(action1)
            return act 

            action = 1
        elif action == 8 :
            dx , dy = (-1 , 0 ) # Move diagonally bottom-left
            action0 = 0
            action1 = 3
           
            act = self.env.step(action0)
            act = self.env.step(action1)
            return act 


        elif action == 9 :        # Move diagonally top-left
            action0 = 1
            action1 = 2
           
            act = self.env.step(action0)
            act = self.env.step(action1)
            return act 

        else:
            raise ValueError(f"Invalid action: {action}")
       
    def action(self, action):
        # Perform any additional processing on the action here
        return action
```

```python
def Policy(state  ) :
    
    if random.uniform(0 ,1 ) < epsilon :
            
            action = env.action_space.sample()
            
    else :  
            action = np.argmax(q_table[state])
            
    return action

def generate_episode():
    
    R  = []   # in this list we maintain the rewards that we gain in our episode
    AS = []   # in this list we maintain the touples that includes Action and State
    done = False 
    state = env.reset()[0]
    action = Policy(state)
    touple = (state , action)
    AS.append(touple)

    while not done :
        act = env.step(action)
#         act[0] ->next_state
#         act[1] ->reward
#         act[2] ->done
        R.append(act[1])
        action = Policy(act[0])
        touple = (act[0] , action)
        AS.append(touple)
        done = act[3]
    R.append(0)
    return R  , AS
          
```

```python
env = gym.make('Taxi-v3')
env = DiagonalTaxiWrapper(env)
q_table = np.zeros([env.observation_space.n , env.action_space.n])
q_table_count = np.zeros([env.observation_space.n , env.action_space.n])

epsilon = 0.1
gamma = 0.9
number_of_episodes = 1000
index = [i for i in range(1000)]

aggregating_rewards = []



for episode in range(number_of_episodes) :
    
    R , AS = generate_episode()
    for state , action in AS :
        first_accurence_index = next(i for i,x in enumerate(AS)  if x[0] == state and x[1] == action)
        G = sum([R[i]*(gamma**i) for i,x in enumerate(AS[first_accurence_index:])])
        
        
        q_table_count[state][action] += 1.0
        
        q_table[state][action] = (q_table[state][action] + G) / q_table_count[state][action]
    aggregating_rewards.append(sum(R))       
    print(f"Episode {episode}")

    
plt.plot(index, aggregating_rewards)
plt.xlabel("episodes")
plt.ylabel("reward")
plt.show()
```
