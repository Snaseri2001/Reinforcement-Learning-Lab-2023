```python
import gym
from gym import Wrapper
import numpy as np
import math
import pygame
import time
import random
```

```python
 
def value_Iteration(env  , gamma , theta ) :
    V = np.zeros(env.observation_space.n)

    while True:      
        delta = 0
        V1 = np.zeros(env.observation_space.n)
    # Iterate over all states    
        for s in range(env.observation_space.n):  
            if s == 5 or s == 7 or s == 11 or s == 12   :
                V1[s] =0
            elif  s == 15:
                V1[s] = 1
            else :
                Q = np.zeros(env.action_space.n)        
                for a in range(env.action_space.n):   
                
                    for t in env.P[s][a]:      
                        reward = 0
                    # Calculate the expected value
                
                        if t[3] == True and t[1]== 15:
                            reward = 1
                        elif t[3] == True :
                            reward = -2
                        elif t[3]==False :
                            reward = 0
                        Q[a] += t[0] * (reward + gamma  *V[t[1]])
        # Update the value function        
                new_V = np.max(Q)        
                delta = max(delta, np.abs(new_V - V[s])) 
                V1[s] = new_V
    # Check for convergence  
            
        V = V1
        print(delta)
        if delta < theta:        
                break

```

```python
class CustomRewardWrapper(gym.Wrapper):
    def __init__(self, env):
        super().__init__(env)

    def step(self, action):
        result = self.env.step(action)
        # modify the reward based on some condition
        if result[1] == 1:
            result[1] = 10
        elif result[1] == 0:
            result[1] = -1
        else:
            result[1] = -2
        return result

    def reset(self):
        return self.env.reset()
```

```python
env = gym.make('FrozenLake-v1' ,  is_slippery=True)


gamma = 0.99
theta = 0.001
value_Iteration(env , gamma  , theta)
```
